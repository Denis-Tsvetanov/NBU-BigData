**Задача 2 (PCA)**. Прикачения dataset съдържа 1000 данни, всяка с 784 характеристики - колони `H1-H784`. Последната колона `y`  дава информация към кой от десетте класа принадлежи даден вектор.

Задачи:

1. Използвайки sklearn намерете броя на главните оси, проекциите на началните вектори върху които дават обща дисперсия (explained variance) не по-малка от 95%.
2. Какъв процент от дисперсията дават проекциите върху първите две главни оси?
3. Всеки елемент от dataset-a нанесете като точка върху двумерна равнина, осите на която са първата и втората главна оси, използвайки подходяща библиотека за графики, напр. matplotlib. Данните от различните класове да са изобразени с различни цветове.




```python
#Based on https://towardsdatascience.com/pca-using-python-scikit-learn-e653f8989e60
#and https://jakevdp.github.io/PythonDataScienceHandbook/05.09-principal-component-analysis.html
```

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

from sklearn import decomposition
from sklearn.preprocessing import StandardScaler

df = pd.read_csv('kontrolno2_data.csv',header=0)
```

```python
#################### 1.Data standardization ######################
```

```python
df
```

```python
x = df.loc[:, df.columns != 'y'].values
y = df.loc[:,['y']].values
x = StandardScaler().fit_transform(x)
```

```python


######################### 2.PCA projection to 2D ####################


```

```python
pca = decomposition.PCA().fit(df)
plt.plot(np.cumsum(pca.explained_variance_ratio_))
plt.xlabel('number of components')
plt.ylabel('cumulative explained variance');
```

```python
#броя на главните оси, проекциите на началните вектори върху които дават обща дисперсия (explained variance) не по-малка от 95%
pca = decomposition.PCA(n_components = 175)
principalComponents = pca.fit_transform(x)
principalDf = pd.DataFrame(data = principalComponents)
finalDf = pd.concat([principalDf, df[['y']]], axis = 1)
finalDf = finalDf.to_numpy()
```

```python
pca.explained_variance_ratio_.sum()
```

```python
#процент от дисперсията дават проекциите върху първите две главни оси
pca.explained_variance_ratio_[0:2]
```

```python



###################### 3. Visualization ###########################


```

```python
#Всеки елемент от dataset-a нанесете като точка върху двумерна равнина, осите на която са първата и втората главна оси
targets = df["y"]
colors = np.random.rand(len(targets))

plt.scatter(principalComponents[:, 0], principalComponents[:, 1],
            c=targets, edgecolor='none', alpha=0.5,
            cmap=plt.cm.get_cmap('prism', 10))
plt.xlabel('component 1')
plt.ylabel('component 2')
plt.colorbar();
```
